## Implementa el algoritmo de la Criba de Erastótenes para encontrar número primos. Encuentra los números primos entre los 10.000 primeros números naturales

La Criba de Eratóstenes es un algoritmo que permite hallar todos los números primos menores que un número natural dado N. Se forma una tabla con todos los números naturales comprendidos entre 2 y N y se van tachando los números que no son primos de la siguiente manera: cuando se encuentra un número entero que no ha sido tachado, ese número es declarado primo, y se procede a tachar todos sus múltiplos

```python
def criba_eratostenes(n):
    # Inicializamos una lista de booleanos para representar los números primos
    # Crea una lista de longitud n+1 con todos los elementos inicializados a True.
    # En este caso, se utiliza para representar los números primos menores o iguales a n
    primos = [True for i in range(n+1)]
    p = 2
    # Iterar sobre todos los números primos menores o iguales a la raíz cuadrada de n.
    while p**2 <= n:
        if primos[p]:
            # Si encontramos un número primo, "tachamos" todos sus múltiplos
            for i in range(p**2, n+1, p):
                primos[i] = False
        p += 1
    # Devolvemos una lista con todos los números primos menores o iguales a n
    return [p for p in range(2, n+1) if primos[p]]

# Imprimimos los números primos menores o iguales a 10.000
print(criba_eratostenes(10000))
```

## Implementa una solución al problema de las 8 reinas en un tablero 8x8

El problema de las ocho reinas es un rompecabezas clásico que involucra un tablero de ajedrez de 8x8 y 8 reinas. El objetivo es colocar las 8 reinas en el tablero de tal manera que ninguna reina amenace a otra, lo que significa que no puede haber dos reinas en la misma fila, columna o diagonal.

Se ha creado la función es_valido, que acepta tres argumentos:

- tablero: matriz numpy que representa el tablero de ajedrez. Un 1 en la posición (i, j) indica que hay una reina en la celda (i, j), y un 0 indica que la celda está vacía.
- fila:el índice de la fila de la celda en la que se quiere colocar una reina
- col: el índice de la columna de la celda en la que quiere colocar una reina.

La función comprueba si hay una reina en la misma fila iterando a través de todas las columnas en la fila dada hasta la columna col, y si encuentra una reina (un 1), devuelve False.

También comprueba si hay una reina en la diagonal superior izquierda o superior derecha, esto se consigue iterando a través de todas las celdas en las diagonales superiores de la celda (fila, col), y si encuentra una reina, devuelve False.


```python
import numpy as np

def es_valido(tablero, fila, col):
    # Comprueba si hay una reina en la misma fila
    for i in range(col):
        if tablero[fila][i] == 1:
            return False
    # Comprueba si hay una reina en la diagonal superior izquierda
    for i, j in zip(range(fila, -1, -1), range(col, -1, -1)):
        if tablero[i][j] == 1:
            return False
    # Comprueba si hay una reina en la diagonal inferior izquierda
    for i, j in zip(range(fila, tablero.shape[0]), range(col, -1, -1)):
        if tablero[i][j] == 1:
            return False
    # Si no hay ninguna reina en la misma fila o en las diagonales superiores o inferiores,
    # entonces la posición es válida y se devuelve True.
    return True
```

Se ha definido resolver_n_reinas_util(tablero, col) para resolver el problema de las N-Reinas de manera recursiva.

La función recibe dos argumentos, tablero y col. Para cada fila en la columna actual, comprueba si es válido colocar una reina en esa posición llamando a la función es_valido(tablero, i, col). Si es válido, coloca una reina en esa posición .

A continuación, hace una llamada recursiva a resolver_n_reinas_util(tablero, col+1) para intentar colocar las reinas restantes en las columnas restantes. Si esta llamada devuelve True, entonces devuelve True inmediatamente, indicando que se ha encontrado una solución.

Si la llamada recursiva devuelve False, significa que no es posible colocar una reina en ninguna posición válida en las columnas restantes, entonces retira la reina de la posición tablero[i][col] al hacer tablero[i][col] = 0, y pasa a la siguiente fila en la columna actual.

En resumen, esta función utiliza la técnica de backtracking para resolver el problema de las N-Reinas. Comienza colocando una reina en la primera posición válida que encuentra en cada columna, y si no puede encontrar una posición válida en una columna dada, vuelve atrás y prueba con las siguientes posiciones.

```python
def resolver_n_reinas_util(tablero, col):
    # Si se han colocado todas las reinas en el tablero sin amenazarse entre sí,
    # entonces devuelve True.
    if col >= tablero.shape[0]:
        return True
    # Para cada fila en la columna actual, comprueba si la posición es válida.
    for i in range(tablero.shape[0]):
        if es_valido(tablero, i, col):
            # Si la posición es válida, coloca una reina en esa posición.
            tablero[i][col] = 1
            # Llama recursivamente a resolver_n_reinas_util() para colocar las reinas restantes.
            if resolver_n_reinas_util(tablero, col+1) == True:
                return True
            # Si no se puede colocar una reina en la columna actual sin amenazarse entre sí,
            # entonces deshace el cambio y prueba con la siguiente fila.
            tablero[i][col] = 0
    # Si no se puede colocar una reina en ninguna fila de la columna actual sin amenazarse entre sí,
    # entonces devuelve False.
    return False
```

Aunque el objetivo es resolver el problema de las 8 Reinas, para experimentar con los costes computacionales en función del número de reinas que se quieran colocar en el tablero, se ha definido la función resolver_n_reinas(n) para resolver el problema de las N-Reinas en un tablero de ajedrez de tamaño n x n.

Para ello se crea un tablero de ajedrez de tamaño n x n y se se llena de ceros para indicar que está vacío. Un 0 en la posición (i, j) del tablero indica que la celda (i, j) está vacía.

Llamando a resolver_n_reinas_util(tablero, 0) se van colocando las reinas en el tablero, comenzando desde la primera columna (col = 0). Si devuelve False, imprime "No existe solución"  Si devuelve True imprime el tablero

```python
def resolver_n_reinas(n):
    # Crea un tablero de ajedrez vacío de tamaño n x n.
    tablero = np.zeros((n,n))
    # Llama a resolver_n_reinas_util() para colocar las reinas en el tablero.
    if resolver_n_reinas_util(tablero, 0) == False:
        # Si no se puede colocar una reina en la primera columna sin amenazarse entre sí,
        # entonces no existe solución.
        print("No existe solución")
        return False
    # Si se han colocado todas las reinas en el tablero sin amenazarse entre sí,
    # entonces imprime el tablero y devuelve True.
    print(tablero)
    return True
```

# Implementa una solución del problema de las Torres de Hanoi para un número arbitrario n de fichas

Divide y vencerás!

Esta técnica consiste en dividir un problema en subproblemas más pequeños y resolver cada subproblema por separado.

En el caso de las Torres de Hanoi, el problema se divide en tres subproblemas:
- mover n-1 discos desde el origen al auxiliar,
- mover el disco n desde el origen al destino y
- mover n-1 discos desde el auxiliar al destino.

Cada uno de estos subproblemas se resuelve recursivamente utilizando la misma función torres_hanoi. La recursión se detiene cuando n es igual a 1 y la función simplemente mueve el disco desde el origen al destino.

```python
# Torres de Hanoy

def torres_hanoi(n, origen, auxiliar, destino):
    ## Si n es igual a 1, mueve el disco desde el origen al destino y termina la función
    if n == 1:
        print("Mover disco 1 desde", origen, "a", destino)
        return

    ## Llama a la función torres_hanoi con n-1 discos desde el origen al auxiliar usando el destino como auxiliar
    torres_hanoi(n-1, origen, destino, auxiliar)

    ## Mueve el disco n desde el origen al destino
    print("Mover disco", n, "desde", origen, "a", destino)

    ## Llama a la función torres_hanoi con n-1 discos desde el auxiliar al destino usando el origen como auxiliar
    torres_hanoi(n-1, auxiliar, origen, destino)

```

# Implemente el código del descenso por gradiente para la regresión lineal.

El modelo de regresión lineal asume una relación lineal entre los datos:

$$ y_i = w_1 x_i + w_0 $$

Un conjunto de datos más realista (con ruido), donde $w_1 = 2$ y $w_0 = 0$:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.uniform(0,1,20)

def f(x): return x*2

noise_variance =0.2
noise = np.random.randn(x.shape[0])*noise_variance
y = f(x) + noise

plt.plot(x, y, 'o', label='y')
plt.plot([0, 1], [f(0), f(1)], 'b-', label='f(x)')
plt.xlabel('$x$', fontsize=15)
plt.ylabel('$y$', fontsize=15)
plt.ylim([0,2])
plt.title('inputs (x) vs targets (y)')
plt.grid()
plt.legend(loc=2)
plt.gcf().set_size_inches((10,6))
plt.show()
```

A continuación se muestra la curva de costo cuadrático en función de diferentes valores de pesos "w" en un modelo de regresión lineal.

Se puede observar cómo cambia el costo a medida que se ajusta el parámetro "w", lo que permite visualizar cómo afecta este parámetro al rendimiento del modelo.

```python
#  La función toma dos argumentos "y" (valores predichos) y "t" (valores reales),
#  calcula el error cuadrático para cada punto y luego suma todos los errores cuadráticos.
def cost(y, t): return ((t - y)**2).sum()

# 100 valores entre 0 y 4
ws = np.linspace(0, 4, num=100)

cost_ws = np.zeros(ws.shape)

#  Para cada valor de "w", se evalúa la función de costo y se almacena el resultado
for ind,w in enumerate(ws):
    cost_ws[ind]=cost(x*w, y)

# Dibuja la curva de costo en función de los valores de w
plt.plot(ws, cost_ws, 'r-')
plt.xlabel('$w$', fontsize=15)
plt.ylabel('Cost', fontsize=15)
plt.title('Cost vs. $w$')
plt.grid()
plt.gcf().set_size_inches((10,6))
plt.show()
```

El algoritmo de descenso de gradiente, es un método de optimización para minimizar la función de costo. A través de varias iteraciones, el algoritmo actualiza el valor de w moviéndose en la dirección del gradiente negativo de la función de costo.

Se ejecuta el algoritmo de descenso de gradiente con un valor inicial de w = 0, una tasa de aprendizaje de 0.1, y durante 100 iteraciones. Luego, se grafican los datos junto con la línea de regresión obtenida después de la optimización. También se crea un gráfico que muestra cómo la función de costo disminuye con cada época.

```python
# Algoritmo de descenso de gradiente
def gradiente(x, y, w):
    return 2 * x * (w*x - y)

def descenso_gradiente(x, y, w_inicial, learning_rate, iteraciones):
    w = w_inicial
    coste = []
    for i in range(iteraciones):
        grad = gradiente(x, y, w)
        w = w - learning_rate * np.mean(grad)
        coste.append(cost(w*x, y))
    return w, coste

# Ejecución del descenso de gradiente e impresión del valor óptimo de w
w_inicial = 0
learning_rate = 0.1
iteraciones = 1000
w_opt, coste = descenso_gradiente(x, y, w_inicial, learning_rate, iteraciones)
print(coste)
# Línea usando el w óptimo
y_pred = w_opt * x

# Gráfica de los datos, la línea de regresión y los costes en cada época
fig, axs = plt.subplots(2, 1, figsize=(10,12))

# Gráfica de los datos y la línea de regresión
axs[0].plot(x, y, 'o', label='y')
axs[0].plot(x, y_pred, 'r-', label='y_pred')
axs[0].plot([0, 1], [f(0), f(1)], 'b-', label='f(x)')
axs[0].set_xlabel('$x$', fontsize=15)
axs[0].set_ylabel('$y$', fontsize=15)
axs[0].set_ylim([0,2])
axs[0].set_title('Entradas (x) vs objetivos (y)')
axs[0].grid(True)
axs[0].legend(loc=2)

# Gráfica de los costes en cada época
axs[1].plot(coste)
axs[1].set_xlabel('Iteraciones', fontsize=15)
axs[1].set_ylabel('Coste', fontsize=15)
axs[1].set_title('Coste vs. Epoca')
axs[1].grid(True)

plt.tight_layout()
plt.show()
```

$$f(x,y)=\sin\left( \frac{1}{2} x^2 - \frac{1}{4} y^2 +3\right)\cos(2x+1-e^y)$$
Para optimizar esta función utilizando el algoritmo de descenso de gradiente, antes se necesita calcular su gradiente. En este caso, se tiene que calcular las derivadas parciales con respecto a
x e y. Las derivadas parciales indicarán la dirección del máximo crecimiento de la función en el espacio de las variables x e y.



```python
from sympy import symbols, sin, cos, exp, diff

# Se definen las variables
x, y = symbols('x y')

# Se define la función
f = sin(1/2 * x**2 - 1/4 * y**2 + 3) * cos(2*x + 1 - exp(y))

# Se calcula la derivada parcial con respecto a x
df_dx = diff(f, x)

# Se calcula la derivada parcial con respecto a y
df_dy = diff(f, y)

print("Derivada parcial con respecto a x:")
print(df_dx)

print("Derivada parcial con respecto a y:")
print(df_dy)

```

```python
from sympy import symbols, sin, cos, exp, diff, lambdify

# Convierte las expresiones de sympy a funciones de python que se pueden evaluar numéricamente
f_conv = lambdify((x, y), f, 'numpy')
df_dx_conv = lambdify((x, y), df_dx, 'numpy')
df_dy_conv = lambdify((x, y), df_dy, 'numpy')

# Define el algoritmo de descenso de gradiente
def descenso_gradiente(x_inicial, y_inicial, learning_rate, iteraciones):
    x = x_inicial
    y = y_inicial
    history = []  # Para guardar la historia de los puntos visitados
    for _ in range(iteraciones):
        grad_x = df_dx_conv(x, y)
        grad_y = df_dy_conv(x, y)
        x -= learning_rate * grad_x
        y -= learning_rate * grad_y
        history.append((x, y, f_conv(x, y)))
    return x, y, history
```

```python
# Seleccionamos un punto inicial aleatorio en el rango [-10, 10] para ambas variables
x_inicial = np.random.uniform(-3, 3)
y_inicial = np.random.uniform(-3, 3)

# Definimos la tasa de aprendizaje y el número de iteraciones
learning_rate = 0.01
iteraciones = 10000

# Ejecutamos el algoritmo de descenso de gradiente
x_final, y_final, history = descenso_gradiente(x_inicial, y_inicial, learning_rate, iteraciones)

# Genera valores de x e y
x_values = np.linspace(-3, 3, 1000)
y_values = np.linspace(-3, 3, 1000)
x_values, y_values = np.meshgrid(x_values, y_values)

# Evalúa la función
z_values = f_conv(x_values, y_values)

# Crea una gráfica de contorno de la función
plt.figure(figsize=(10, 6))
contour = plt.contour(x_values, y_values, z_values, levels=100)
plt.colorbar(contour)
plt.title('Contour plot of the function and gradient descent steps')
plt.xlabel('x')
plt.ylabel('y')

# Extrae los puntos de la historia del descenso de gradiente
x_history, y_history, z_history = zip(*history)

# Superpone los puntos del descenso de gradiente en la gráfica
plt.plot(x_history[:-1], y_history[:-1], 'ro-', markersize=3, linewidth=0.5, label='Iteraciones anteriores')
plt.plot(x_history[-1], y_history[-1], 'bo', markersize=5, label='Última iteración')

plt.legend()  # Agrega una leyenda para identificar los puntos
plt.show()

print(f"El punto inicial es ({x_inicial}, {y_inicial})")
print(f"El punto final después de {iteraciones} iteraciones es ({x_final}, {y_final})")
print(f"El valor de la función en el punto final es {history[-1][-1]}")
```
